```python
# import libraries 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import os
from pycaret.classification import *
```

```python
# change working dir to top of work directory 
os.chdir('/home/rstudio/work')
```

```python
# read in data and view the first few lines 
data_all_cols = pd.read_csv('data/source_data/petfinder_train_dataset.csv')

pd.set_option('display.max_columns', None)
data_all_cols.head()
```

```python
# keep only columns that make sense to use in the model 
data = data_all_cols.drop(['Name', 'Description','PetID'], axis=1)

# List of categorical columns so make it clear which need to be one hot encoded during set up
categorical_columns = [
    'Type', 'Breed1', 'Gender', 'Color1', 'Color2',
    'MaturitySize', 'FurLength', 'Vaccinated', 'Dewormed',
    'Sterilized', 'Health', 'State', 'RescuerID','Breed2','Color3'
]

# Convert specified columns to 'category'
data[categorical_columns] = data[categorical_columns].astype('category')
```

```python
print(data.dtypes)
```

```python
# split data so that we can train on some and predict on the rest
data_train = data.sample(frac=0.9, random_state=786).reset_index(drop=True)
data_predict = data.drop(data_train.index).reset_index(drop=True)

data_train.to_csv('data/working_data/filtered_transformed_training_data.csv', sep = ',', header = True, index = False)
data_predict.to_csv('data/working_data/filtered_transformed_prediction_data.csv', sep = ',', header = True, index = False)

print('Data for modeling:' + str(data_train.shape))
print('Data for predictions:' + str(data_predict.shape))
```

```python
# initialize the environment in pycaret
# creates transformation pipeline to prepare the data for modeling and deployment
s = setup(data_train, target = 'AdoptionSpeed', session_id = 123)
```

```python
best = compare_models(exclude=['lightgbm'])
```

```python
# given that none of these models are performing well, let's figure out which features are important

# build the extra trees classifier to look at important features 
et = create_model('et')
plot_model(et, plot='feature')
plt.savefig('results/figures/et_model_top_features_importance_nofeature_selection_plot.png', dpi=300, bbox_inches='tight')
```

```python
# Retrieve feature names from PyCaret's transformed pipeline
feature_names = get_config('X_train_transformed').columns

# Print the total number of features
print(f"Total number of features: {len(feature_names)}")

# Retrieve feature importances
importances = et.feature_importances_

# Combine feature names with importances
et_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Identify features to keep
important_features = et_importance[et_importance['Importance'] > 0.01]['Feature'].tolist()
print(f"Number of important features: {len(important_features)}")

# Retrieve the transformed dataset from PyCaret
X_train_transformed = get_config('X_train_transformed')
data_train_transformed_filtered = X_train_transformed[important_features]  # Subset transformed data

# Add the target column back to the transformed dataset
data_train_filtered = pd.concat(
    [data_train_transformed_filtered, data_train[['AdoptionSpeed']].reset_index(drop=True)],
    axis=1
)

data_train_filtered.to_csv('data/working_data/filtered_transformed_top_features_train_data.csv', sep = ',', header = True, index = False)
print(f"Shape of filtered dataset: {data_train_filtered.shape}")
```

```python
# re-complete set up with this new dataset 
s = setup(data_train_filtered, target = 'AdoptionSpeed', session_id = 123)

best_filtered = compare_models(exclude=['lightgbm'])
```

```python
# modify with held data to only have the same top 25 important features 
# Retrieve the transformed data pipeline using config and apply to prediction data
X_predict_transformed = get_config('X_train_transformed')

# Filter the transformed prediction data to only include the important features
data_predict_transformed_filtered = X_predict_transformed[important_features]

# add the target (adoption speed) column back in 
data_predict_filtered_with_target = pd.concat(
    [data_predict_transformed_filtered, data_predict[['AdoptionSpeed']].reset_index(drop=True)],
    axis=1
)

# Save the transformed prediction data (optional)
data_predict_filtered_with_target.to_csv('data/working_data/filtered_transformed_top_features_predict_data.csv', sep=',', header=True, index=False)

# Print the shape of the filtered prediction dataset
print(f"Shape of filtered prediction dataset: {data_predict_filtered_with_target.shape}")
```
