```python
# import libraries 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import os
from pycaret.classification import *
import matplotlib as plt
```

```python
# change working dir to top of work directory 
os.chdir('/home/rstudio/work')
```

```python
# read in data and view the first few lines 
data_all_cols = pd.read_csv('/home/rstudio/work/data/source_data/petfinder_train_dataset.csv')

pd.set_option('display.max_columns', None)
data_all_cols.head()
```

```python
# keep only columns that make sense to use in the model 
data = data_all_cols.drop(['Name', 'Description','PetID','RescuerID'], axis=1)

# List of categorical columns so make it clear which need to be one hot encoded during set up
categorical_columns = [
    'Type', 'Breed1', 'Gender', 'Color1', 'Color2',
    'MaturitySize', 'FurLength', 'Vaccinated', 'Dewormed',
    'Sterilized', 'Health', 'State','Breed2','Color3'
]

# Convert specified columns to 'category'
data[categorical_columns] = data[categorical_columns].astype('category')
```

```python
print(data.dtypes)
```

```python
# split data so that we can train on some and predict on the rest
data_train = data.sample(frac=0.9, random_state=786).reset_index(drop=True)
data_predict = data.drop(data_train.index).reset_index(drop=True)

data_train.to_csv('/home/rstudio/work/data/working_data/filtered_transformed_training_data.csv', sep = ',', header = True, index = False)
data_predict.to_csv('/home/rstudio/work/data/working_data/filtered_transformed_prediction_data.csv', sep = ',', header = True, index = False)

print('Data for modeling:' + str(data_train.shape))
print('Data for predictions:' + str(data_predict.shape))
```

```python
# initialize the environment in pycaret
# creates transformation pipeline to prepare the data for modeling and deployment
s = setup(data_train, target = 'AdoptionSpeed', session_id = 123)

data_train_transformed = get_config('X_train_transformed')
print(data_train_transformed.isnull().sum())  # Check for NaNs in the transformed data
```

```python
best = compare_models(exclude=['lightgbm'])
```

```python
# given that none of these models are not performing well, let's figure out which features are important

# build the extra trees classifier to look at important features 
et = create_model('et')
plot_model(et, plot='feature')
```

```python
# Retrieve feature names from PyCaret's transformed pipeline
feature_names = get_config('X_train_transformed').columns

# Print the total number of features
print(f"Total number of features: {len(feature_names)}")

# Retrieve feature importances
importances = et.feature_importances_

# Combine feature names with importances
et_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Identify features to keep
important_features = et_importance[et_importance['Importance'] > 0.01]['Feature'].tolist()
print(f"Number of important features: {len(important_features)}")

# Retrieve the transformed dataset from PyCaret
X_train_transformed = get_config('X_train_transformed')
data_train_transformed_filtered = X_train_transformed[important_features]  # Subset transformed data

# Add the target column back to the transformed dataset
data_train_filtered = pd.concat(
    [data_train_transformed_filtered, data_train[['AdoptionSpeed']].reset_index(drop=True)],
    axis=1
)

data_train_filtered.to_csv('/home/rstudio/work/data/working_data/filtered_transformed_top_features_train_data.csv', sep = ',', header = True, index = False)
print(f"Shape of filtered dataset: {data_train_filtered.shape}")
```

```python
# re-complete set up with this new dataset 
s = setup(data_train_filtered, target = 'AdoptionSpeed', session_id = 123)
# we did improve our accuracy, so we will use these features for our models 
best_filtered = compare_models(exclude=['lightgbm'])
```

```python
# Initialize the setup without the split
s = setup(data=data_predict, 
          target='AdoptionSpeed', 
          session_id=123)

# Retrieve the transformed train dataset (this includes the target column in the transformed data)
X_train_transformed = get_config('X_train_transformed')

# Now, retrieve the 'AdoptionSpeed' column from the transformed training data
y_train_transformed = get_config('y_train')
y_train_transformed_df = pd.DataFrame(y_train_transformed, columns=['AdoptionSpeed'])

# Filter the transformed features based on the important features you selected
data_pred_transformed_filtered = X_train_transformed[important_features]

# Concatenate the target column from the transformed train data to the filtered prediction dataset
data_pred_transformed_filtered = pd.concat(
    [X_train_transformed, y_train_transformed_df],  # Add the target column
    axis=1
)

# Print the shape of the final dataset
print(f"Shape of filtered prediction dataset: {data_pred_transformed_filtered.shape}")
```
